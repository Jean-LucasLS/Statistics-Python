```python
import pandas as pd

FILE_PATH = 'dados.csv'

df = pd.read_csv(FILE_PATH)
df.head(2)
```

```python
df['Renda'].quantile([0.25, 0.5, 0.75])
```

```python
df['Renda'].quantile([i/10 for i in range(1, 10)])
```

```python
import seaborn as sns

ax = sns.distplot(df['Idade'], hist_kws={'cumulative': True}, kde_kws={'cumulative': True}, bins=10)
ax.figure.set_size_inches(12, 6)
ax.set_title('Distribuição de Frequências Acumuladas', fontsize=18)
ax.set_ylabel('Acumulado', fontsize=12)
ax.set_xlabel('Anos', fontsize=12)
ax
```
